```python
# Cross Examine
pd.options.plotting.backend = 'plotly'

variable_labels = {"SP.POP.DPND.OL": "Old Age Dependency Ratio",
                   "SP.POP.DPND.YG": "Young Age Dependency Ratio", 
                   "SP.POP.DPND": "Total Age Dependency Ratio"}

countries = {"JPN":"Japan",
             "NER":"Niger",
             "USA":"United States", 
             "QAT":"Qatar", 
             "BRA":"Brazil"
            }


df = wbdata.get_dataframe(variable_labels, country=countries, parse_dates=True)
df = df.unstack('country')

comparison_df = df.loc[:, [
    ('Total Age Dependency Ratio', 'Japan'), 
    ('Total Age Dependency Ratio', 'Niger'),
    ('Total Age Dependency Ratio', 'United States'),
    ('Total Age Dependency Ratio', 'Qatar'), 
    ('Total Age Dependency Ratio', 'Brazil')
]]

comparison_df.index = comparison_df.index.year
comparison_df.columns = ['Japan', 'Niger', 'United States', 'Qatar', 'Brazil'] # Rename for the legend

fig = comparison_df.plot(
    title="Cross Examine: Japan and Niger's Total Age Dependency Ratio vs. Other Countries",
    labels={'value': 'Ratio (%)', 'date': 'Year'}
)

fig.show()
```

```python
# Cross Examine
variable_labels = {"SP.POP.DPND.OL": "Old Age Dependency Ratio",
                   "SP.POP.DPND.YG": "Young Age Dependency Ratio"}

df = wbdata.get_dataframe(variable_labels, country=countries, parse_dates=True)
df = df.unstack('country')

comparison_df = df.loc[:, [
    ('Old Age Dependency Ratio', 'Japan'), 
    ('Young Age Dependency Ratio', 'Niger')
]]

comparison_df.index = comparison_df.index.year
comparison_df.columns = ['Japan (Old Age)', 'Niger (Young Age)'] # Rename for the legend

fig = comparison_df.plot(
    title="Cross Examine: Japan's Old Dependency Ratio vs. Niger's Young Dependency Ratio",
    labels={'value': 'Ratio (%)', 'date': 'Year'}
)

fig.show()
```
