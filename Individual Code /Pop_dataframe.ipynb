## **A Snapshot of Japan's Young/Old Age Distribiution Over Time**

```python
pd.set_option('display.float_format', '{:.3f}'.format)

row_indices_1 = [("Japan", "2024-01-01"), ("Japan", "2014-01-01"), ("Japan", "2004-01-01"), ("Japan", "1994-01-01"), ("Japan", "1984-01-01"), ("Japan", "1974-01-01"), ("Japan", "1960-01-01")]
japan_time = data_frame.loc[row_indices_1, ["Pop. all ages total","Pop. 0-14 total", "% of total Pop. 0-14", "Pop. 65 and above total", "% of total Pop. 65 and above", "Age dependency ratio"]]

japan_time.style \
  .format({
      'Pop. all ages total': '{:.0f}',     
      'Pop. 0-14 total': '{:.0f}',    
      '% of total Pop. 0-14': '{:.3f}%',      
      'Pop. 65 and above total': '{:.0f}',     
      '% of total Pop. 65 and above': '{:.3f}%', 
      'Age dependency ratio': '{:.3f}'
  }) \
  .set_properties(**{'text-align': 'right', 'padding': '5px'}) \
  .set_table_styles([
      {'selector': 'thead th', 'props': [('background-color', '#40466e'),
                                          ('color', 'white'),
                                          ('text-align', 'center'),
                                          ('font-weight', 'bold')]}
  ])
```
